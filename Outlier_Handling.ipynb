```python
def quantile_operation(outlier_df, column):
    quantile_1 = outlier_df[column].quantile(0.25)
    quantile_3 = outlier_df[column].quantile(0.75)
    iqr = quantile_3 - quantile_1
    Lower_tail = quantile_1 - 1.5 * iqr
    Upper_tail = quantile_3 + 1.5 * iqr

    return Lower_tail, Upper_tail
```

```python
#method : scaling, log transformation, cube Root normalization, Box-Cox transformation
from scipy.stats import skew 
def outlier(value, Lower_tail, Upper_tail):
    return ((value > Upper_tail) or (value < Lower_tail))
outlier_columns = ['BMI', 'PhysicalHealth', 'MentalHealth', 'SleepTime']
for column in outlier_columns:
    print('이상치 적용하기 전 ' + column + ' 의 skewness : ', skew(outlier_df[column]))

    Lower_tail, Upper_tail = quantile_operation(outlier_df, column)
    outlier_df[column] = outlier_df[column].apply(lambda x : outlier_df[column].median() if (x > Upper_tail) or (x < Lower_tail) else x)

    Lower_tail, Upper_tail = quantile_operation(outlier_df, column)
    print('이상치 적용한 후 ' + column + '의 skewness : ', skew(outlier_df[column]))
```
